```python
import numpy as np

# Training data: [feature_1, feature_2]
X_train = np.array([
    [1.0, 2.0],
    [1.5, 1.8],
    [2.0, 2.5],
    [6.0, 8.0],
    [7.0, 8.5],
    [8.0, 7.5],
    [3.0, 3.5],
    [3.5, 4.0],
    [9.0, 9.0]
])

# Labels/classes
y_train = np.array([
    0, 0, 0,
    1, 1, 1,
    0, 0,
    1
])

# Test data
X_test = np.array([
    [1.2, 2.1],
    [7.5, 8.0],
    [3.2, 3.8],
    [8.5, 8.7],
    [2.5, 3.0]
])

# Expected rough predictions:
# [0, 1, 0, 1, 0]
```

```python
import numpy as np
import matplotlib.pyplot as plt
import statistics as st
class OKNN:
    def __init__(self,X,y,k=5):
        self.X=np.array(X)
        y=np.array(y)
        self.y=np.array(y).reshape(1,y.shape[0])
        self.k=k
    def euc_norm(self,a,b):
        return np.sqrt(np.sum((a-b)**2))
    def fit(self):
        return 'Model was fit'
    def predict(self,X_pred):
            X_pred=np.array(X_pred)
            distance=np.ones(shape=(1,self.X.shape[0]))
            for i in range(0,self.X.shape[0],1):
                distance[0][i]=self.euc_norm(X_pred,self.X[i,:])
            distance=np.array(distance)
            distance=distance.reshape(1,distance.shape[1])
            dist_and_y=np.vstack([distance,self.y])
            i=np.argsort(dist_and_y[0])
            dist_and_y=dist_and_y[:,i]
            return st.mode(dist_and_y[1,:self.k])
    def decision_boundary(self):
        plt.scatter(self.X[0],self.X[1])
        xx,yy=np.meshgrid(np.linspace(self.X[0].min(),self.X[0].max()),np.linspace(self.X[1].min(),self.X[1].max()))
        grid=np.vstack([xx.ravel(),yy.ravel()])
        for j in range(0,xx.ravel().shape[0]):
            z=self.predict(grid.T[j,:])
            plt.contour(xx[j],yy[j],z.reshape(xx.shape))
        
```

```python
oknn=OKNN(X_train,y_train)
oknn.fit()
oknn.decision_boundary()
```
